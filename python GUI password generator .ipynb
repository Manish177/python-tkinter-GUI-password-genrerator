```python
from tkinter import *
import random
root=Tk()
root.title("Password genrator")
global password

def pass_gen():
    chars=''
    chars=''.join(chars_all)
    global password
    password=''
    pass_len=int(myentry.get())    
    if not isinstance(pass_len,int):
        myentry.delete(0,END)
    for i in range(int(pass_len)):
        pass_char=random.choice(chars)
        password=password+pass_char
    label1=Label(root, text=password)
    label1.grid(row=3, column=1)

chars_all=[]
num='1234567890'
char_small='qwertyuiopasdfghjklzxcvbnm'
char_capital='QWERTYUIOPASDFGHJKLZXCVBNM'
char_special='!@#$%^&*()_-+[]{};:,./<>?'


cb1 = IntVar()
cb2 = IntVar()
cb3 = IntVar()
cb4 = IntVar()

def isChecked():  
    if cb1.get()==1:
        if not chars_all.count(num):
            chars_all.append(num)
    else:
        if chars_all.count(num):
            chars_all.remove(num)
        
    if cb2.get()==1:
        if not chars_all.count(char_small):
            chars_all.append(char_small)
    else:
        if chars_all.count(char_small):
            chars_all.remove(char_small)
        
    if cb3.get()==1:
        if not chars_all.count(char_capital):
            chars_all.append(char_capital)
    else:
        if chars_all.count(char_capital):
            chars_all.remove(char_capital)
        
    if cb4.get()==1:
        if not chars_all.count(char_special):
            chars_all.append(char_special)
    else:
        if chars_all.count(char_special):
            chars_all.remove(char_special)
            
def copyto():
    root.clipboard_clear()
    root.clipboard_append(password)

    
cbutton1=Checkbutton(root,text='numbers', variable=cb1,onvalue=1,
                     offvalue=0,command=isChecked)
cbutton1.grid(row=0,column=0, sticky=W, padx=20)

cbutton2=Checkbutton(root,text='charcters small', variable=cb2,onvalue=1,
                     offvalue=0,command=isChecked)
cbutton2.grid(row=1,column=0,sticky=W,padx=20)

cbutton3=Checkbutton(root,text='charcters capital',variable=cb3,onvalue=1,
                     offvalue=0,command=isChecked)
cbutton3.grid(row=2,column=0,sticky=W,padx=20)

cbutton4=Checkbutton(root,text='Special characters',variable=cb4,onvalue=1,
                     offvalue=0,command=isChecked)
cbutton4.grid(row=3,column=0,sticky=W,padx=20)

myentry=Entry(root, width=30)
myentry.grid(row=1, column=1, sticky=E, padx=10)

label1=Label(root, text='Enter Password length in the box')
label1.grid(row=0, column=1, sticky=E, padx=10)

btn=Button(root, width=22, text='click to generate a password', command=lambda:pass_gen())
btn.grid(row=2, column=1, padx=10)

btn=Button(root, width=22, text='Copy to clipboard', command=lambda:copyto())
btn.grid(row=4, column=1, padx=10)

root.mainloop()
```
